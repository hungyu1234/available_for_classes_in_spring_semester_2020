```python
import pandas as pd
```

```python
def make_df(data):
    df = pd.DataFrame(index = range(len(data.index) // 10)) 
    df['강의번호'] = data['data'][4::10].values
    df['강의명'] = data['data'][5::10].values
    df['학점-시간'] = data['data'][6::10].values
    df['담당교수'] = data['data'][7::10].values
    df['강의시간'] = data['data'][8::10].values
    df['비고'] = data['data'][9::10].values
    df.set_index(df['강의명'], inplace = True)
    del df['강의명']
    
    print('총 강의 수:', len(df))
    return df
```

```python
def imposible(df):
    list = ['지식경영학부','간호학과', '외국인', '인문', '이공', '중국인']
    for i in list:
        df['X'] = df['비고'].apply(lambda x: 1 if i in x else 0)
        df = df[df['X'].isin([1]) == False]
        del df['X']
        
    return df
```

```python
def time(df):
    list = ['월3,4,5','월4,5,6', '월5,6,7', '월6,7,8', '월7,8,9','월8,9,10',
             '월4,5', '월5,6', '월6,7', '월7,8', '월8,9',
             '화1,2,3', '화2,3,4','화3,4,5','화4,5,6', '화5,6,7', '화6,7,8', '화7,8,9', '화8,9,10',
             '화1,2', '화2,3', '화3,4', '화4,5','화5,6','화6,7','화7,8', '화8,9',
             '수1,2,3', '수2,3,4', '수3,4,5', '수4,5,6', '수5,6,7', '수6,7,8',
             '수1,2', '수2,3', '수4,5', '수5,6', '수6,7',
             '목5,6,7', '목6,7,8', '목7,8,9',
             '목6,7', '목7,8', '목8,9'
            ]
    for i in list:
        df['X'] = df['강의시간'].apply(lambda x: 1 if i in x else 0)
        df = df[df['X'].isin([1]) == False]
        del df['X']
        
    return df
```

```python
def free_friday(df):
    df['X'] = df['강의시간'].apply(lambda x: 1 if '금' in x else 0)
    df = df[df['X'].isin([1]) == False]
    del df['X']
    
    return df
```

```python
def free_morning(df):
    list = ['월1', '화1', '수1', '목1', '금1']
    
    for i in list:
        df['X'] = df['강의시간'].apply(lambda x: 1 if i in x else 0)
        df = df[df['X'].isin([1]) == False]
        del df['X']
        
    return df
```

```python
#main영역
data = pd.read_csv('강의시간표.csv')
df = make_df(data)
df = imposible(df)
df = time(df)
df = free_friday(df)
df = free_morning(df)

print('수강 가능한 강의 수:', len(df))
df
```

```python
writer = pd.ExcelWriter('결과물.xlsx')
df.to_excel(writer, 'Sheet1')
writer.save()
```
